```python
import cv2
import numpy as np
from tensorflow import keras

# Load your CNN model
model = keras.models.load_model(r"C:\Users\Dell\Downloads\square_rectangle_model.keras")
print("âœ… Model loaded successfully")

# Label mapping (adjust if needed)
idx2label = {0: "rectangle", 1: "square"}

# Resize with padding function
def resize_with_padding(img, target_size=(150, 150)):
    h, w = img.shape[:2]
    th, tw = target_size
    scale = min(tw / w, th / h)
    nw, nh = int(w * scale), int(h * scale)
    img_resized = cv2.resize(img, (nw, nh))
    top = (th - nh) // 2
    bottom = th - nh - top
    left = (tw - nw) // 2
    right = tw - nw - left
    img_padded = cv2.copyMakeBorder(
        img_resized, top, bottom, left, right,
        cv2.BORDER_CONSTANT, value=[0, 0, 0]
    )
    return img_padded

# Start webcam
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    print("âŒ Error: Could not open webcam.")
    exit()

print("ğŸ“¸ Press 'q' to quit")

while True:
    ret, frame = cap.read()
    if not ret:
        print("âŒ Failed to grab frame")
        break

    # Convert and resize for CNN model
    img_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    img_resized = resize_with_padding(img_rgb, (150, 150))
    img_array = np.expand_dims(img_resized / 255.0, axis=0)

    # Predict
    preds = model.predict(img_array)
    class_index = np.argmax(preds)
    confidence = np.max(preds) * 100
    label = idx2label[class_index]

    # Display prediction on webcam frame
    text = f"{label} ({confidence:.2f}%)"
    cv2.putText(frame, text, (30, 40),
                cv2.FONT_HERSHEY_SIMPLEX, 1,
                (0, 255, 0), 2)

    cv2.imshow("Rectangle vs Square Classifier", frame)

    # Quit when 'q' pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
